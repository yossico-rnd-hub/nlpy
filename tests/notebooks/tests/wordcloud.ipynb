```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline

import spacy
nlp = spacy.load('en')

# Read the text.
file = '../../docs/text6'
with open(file, 'r') as f:
    text = f.read()
    doc = nlp(text)
    string = ' '.join(t.lemma_ for t in doc if not t.is_punct)

stopwords = set(STOPWORDS)
stopwords.update(["say"])

# Generate a word cloud image
# (lower max_font_size)
wordcloud = WordCloud(
    max_font_size=50,
     max_words=1000, 
    background_color="white",
    stopwords=stopwords,
).generate(string)

# Display the generated image:
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
